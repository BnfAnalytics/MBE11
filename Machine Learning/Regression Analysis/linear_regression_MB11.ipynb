```python
def change_size(size):
  if 'M'in size:
    x=size[:-1] #19M
    x=float(x)*1000
    return x
  if 'k' in size:
    x=size[:-1] #19k
    x=float(x)
    return x
  else:
    return None
```

# Preprocessing of Data

```python
# Convert categorical data into numerical data
mbe11=pd.get_dummies(mb11)
```

```python
mbe11.shape
```

# Split data into dependent(y) and independent variables(X)

```python
mbe11.head()
```

```python
y=mbe11.pop('Rating') # dependent variable
x=mbe11 # independent variable
```

```python
y
```

# Separate the data into X_train,X_test,y_train and y_test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# Train the Linear Regression model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)  # "y_test" is actual output and "y_pred" is the predicted outputs
```

# Evaluation of Model
1. R2 score
2. Mean Squared Error

```python
from sklearn.metrics import r2_score,mean_squared_error
print("R2 Score:",r2_score(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
```

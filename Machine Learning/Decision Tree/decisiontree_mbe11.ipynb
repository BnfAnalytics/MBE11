```python
print(emp.EmployeeCount.unique())
print(emp.Over18.unique())
print(emp.StandardHours.unique())
print(emp.EmployeeNumber.unique())
```

```python
emp=emp.drop(['EmployeeCount','Over18','StandardHours','EmployeeNumber'],axis=1)
```

```python
emp.shape
```

```python
# check the values in categorical data/features
print(emp.BusinessTravel.value_counts())
print('*************************************')
print(emp.Department.value_counts())
print('*************************************')
print(emp.EducationField.value_counts())
print('*************************************')
print(emp.Gender.value_counts())
print('*************************************')
print(emp.JobRole.value_counts())
print('*************************************')
print(emp.MaritalStatus.value_counts())
```

```python
# Encoding categorical data /variables into numerical variables/data
emp_dummies=pd.get_dummies(emp[['BusinessTravel','Department','EducationField','Gender',
                                'JobRole','MaritalStatus','OverTime']],drop_first=True)
```

```python
emp_dummies.head()
```

```python
emp=emp.drop(['BusinessTravel','Department','EducationField','Gender',
                                'JobRole','MaritalStatus','OverTime'],axis=1)
```

```python
emp=pd.concat([emp,emp_dummies],axis=1)
```

```python
emp.head()
```

```python
x=emp.drop('Attrition',axis=1)
y=emp['Attrition']
```

```python
# Split data into train ad test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=21)
```

```python
# Object-1: Implementation of Decision Tree
from sklearn import tree
dt1=tree.DecisionTreeClassifier()
dt1.fit(x_train,y_train)
y_pred1=dt1.predict(x_test)
```

```python
# Object-2
dt2=tree.DecisionTreeClassifier(criterion='entropy',max_depth=4)
dt2.fit(x_train,y_train)
y_pred2=dt2.predict(x_test)
```

```python
# Object-3
dt3=tree.DecisionTreeClassifier(criterion='entropy',max_depth=2,min_samples_leaf=2)
dt3.fit(x_train,y_train)
y_pred3=dt3.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred1))
print(metrics.accuracy_score(y_test,y_pred2))
print(metrics.accuracy_score(y_test,y_pred3))
```

```python
# select hyperparameters using GRidsearch
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['gini','entropy'],
            "max_features":[1,3,5,10],
            "min_samples_split":[2,3,5,10],
            "min_samples_leaf":[1,2,3,4,5],
            "max_depth":[2,3,4,5]}
grid=GridSearchCV(dt1,param_grid=param_grid,cv=5)
grid.fit(x_train,y_train)
print(grid.best_params_)
```

```python
# Object-4:With Gridsearch
dt4=tree.DecisionTreeClassifier(criterion='entropy',max_depth=2,max_features=5,min_samples_leaf=2,min_samples_split=3)
dt4.fit(x_train,y_train)
y_pred4=dt4.predict(x_test)
```

```python
print(metrics.accuracy_score(y_test,y_pred4))
```

```python
# plot decision tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(40,20))
plot_tree(dt4,class_names=['No','Yes'],feature_names=x.columns,filled=True)
plt.show()
```

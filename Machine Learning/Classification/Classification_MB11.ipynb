```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=lr.iloc[:,[1,2,3]].values
y=lr.iloc[:,4].values
```

```python
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,0]=le.fit_transform(X[:,0])
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(X_train)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
print(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
# Create Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
print('Accuracy Score: ',accuracy_score(y_test,y_pred))
print('Precision Score: ',precision_score(y_test,y_pred))
print('Recall Score: ',recall_score(y_test,y_pred))
print('F1 Score: ',f1_score(y_test,y_pred))
```

# Implement Gaussian NB

```python
from sklearn.naive_bayes import GaussianNB
classifier_NB=GaussianNB()
classifier_NB.fit(X_train,y_train)
```

```python
y_predNB=classifier_NB.predict(X_test)
y_predNB
```

```python
cmNB=confusion_matrix(y_test,y_predNB)
cmNB
```

```python
print('Accuracy Score  NB: ',accuracy_score(y_test,y_predNB))
print('Precision Score NB: ',precision_score(y_test,y_predNB))
print('Recall Score NB: ',recall_score(y_test,y_predNB))
print('F1 Score NB: ',f1_score(y_test,y_predNB))
```
